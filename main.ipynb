```python
# Read the routes to count from the file
def get_routes():
    routes = system.getRoutes()
    routeNames = open('new_brunswick.txt').read().split('\n')

    routesNB = []

    for route in routes:
        if route.nameOrig in routeNames:
            routesNB.append(route)
    
    return routesNB
```

```python
routes = get_routes()
for route in routes:
    print(route.nameOrig)
    print(route.myid)
```

```python
# get new brunswick stops

# 

def get_stops():
    nb_stops = []
    routes = get_routes()
    stops = system.getStops()

    route_ids = set(str(route.myid) for route in routes)

    for stop in stops: 
        if list(stop.routesAndPositions.keys())[0] in route_ids:
            nb_stops.append(stop)
    
    return nb_stops
stops = get_stops()
stops
```

```python
def printAny(object):
    for attr in dir(object):
        if not attr.startswith('_'):
            print(attr, getattr(object, attr))
```

```python
stops = get_stops()

# create a json file with the stops
import json

stops_json = []

for stop in stops:
    stops_json.append({
        'name': stop.name,
        'id': stop.id,
        'latitude': stop.latitude,
        'longitude': stop.longitude,
        'routes': stop.routesAndPositions,
        'radius': stop.radius
    })

with open('stops.json', 'w') as f:
    json.dump(stops_json, f)
```

```python
routes = get_routes()

# create a json file with the routes
import json

routes_json = []

""" getStops <bound method Route.getStops of <passiogo.Route object at 0x000002254D5DF280>>
goPrefixRouteName 1
goShowSchedule 0
groupColor None
groupId None
id 10098
latitude 40.734819000
longitude -74.164721000
mapApp 1
myid 4098
name PennExpr Penn Station Express
nameOrig Penn Station Express
outdated 1
serviceTime is not provided: no bus on the route
serviceTimeShort No bus in service
shortName PennExpr
system <passiogo.TransportationSystem object at 0x000002254D5C0FA0>
systemId 1268 """

for route in routes:
    routes_json.append({
        'name': route.name,
        'id': route.id,
        'myid': route.myid,
        'latitude': route.latitude,
        'longitude': route.longitude,
        'stops': [stop.id for stop in route.getStops()],
        'distance': route.distance,
        'fullname': route.nameOrig,
        'shortName': route.shortName,
    })

with open('routes.json', 'w') as f:
    json.dump(routes_json, f)
```
