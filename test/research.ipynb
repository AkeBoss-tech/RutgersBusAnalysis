```python
import requests
from bs4 import BeautifulSoup

# Fetch the HTML content from the URL
url = 'https://oirap.rutgers.edu/ResearchCenters/'
""" 
response = requests.get(url) """

with open('center.html', 'r', encoding='utf-8') as file:
    response = file.read()
# Check if the request was successful

soup = BeautifulSoup(response, 'html.parser')

# Find the content div id (modify if needed)
content_div = soup.find('div', id='centerItems')

# Extract all links that start with 'center.aspx'
links = content_div.find_all('a', href=True)
print(f'Found {len(links)} links on the page')

center_links = []
for link in links:
    if link['href'].startswith('center.aspx'):
        center_links.append(f'{link["href"]}')

# Store the extracted links in a file
with open('center_links.txt', 'w', encoding='utf-8') as file:
    for item in center_links:
        file.write(f'{item}\n')

print('Links have been successfully saved to center_links.txt')
```

```python
import requests
from bs4 import BeautifulSoup

# Function to scrape data from each center page
def scrape_center_info(num):
    url = f"https://oirap.rutgers.edu/ResearchCenters/handlers/GetResearchCenter.ashx?id={num}&_=1728741338590"
    
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to fetch {url}")
        return None
    
    # store it in a file
    with open(f'center_{num}.json', 'w', encoding='utf-8') as file:
        file.write(response.text)
    
    return response.text

# Function to read the links from the file
def process_links(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        number = [line.strip().split('id=')[1] for line in file.readlines()]
    
    all_data = []

    # Iterate through each link and scrape the data
    for link in number:
        full_url = link
        print(f"Scraping {full_url}...")
        data = scrape_center_info(full_url)
        if data:
            all_data.append(data)

    return all_data

# Save the scraped data to a file
def save_data_to_file(data, output_file):
    with open(output_file, 'w', encoding='utf-8') as file:
        for item in data:
            file.write(f"{item}\n")

# Main function to control the process
def main():
    input_file = 'center_links.txt'
    output_file = 'center_data.txt'
    
    # Process the links and get the scraped data
    scraped_data = process_links(input_file)
    
    # Save the scraped data to a file
    save_data_to_file(scraped_data, output_file)
    print(f"Scraped data saved to {output_file}")
```

```python
main()
```
